## Train Test Split

As the size of data is not very large we are using 60%(training) to 40%(testing) split as discussed in Tutorial session.

```python
# Separating the expected output from features

y = main_df["MEDV"].to_numpy()
main_df.drop(axis="columns", labels="MEDV", inplace=True)
X = main_df.to_numpy()
# print(X)
# print(y)
```

```python
def train_test_split(X,y,training_size):

    np.random.seed(0)
    random_order = np.random.permutation(X.shape[0])

    # Below given is using random_order to randomly order the rows exactly in same order in both X and y
    X = X[random_order]
    y = y[random_order]

    division = int(training_size*X.shape[0])

    train_x = X[0:division,:]
    train_y = y[0:division]
    test_x = X[division:]
    test_y = y[division:]
    
    return (train_x,train_y,test_x,test_y)
```

```python
train_x,train_y,test_x,test_y = train_test_split(X,y,0.6)
```

```python
#function to normalise data before finding gradient
def normalize(train_x,test_x):
    
    mean = []
    std = []

    for i in range(train_x.shape[1]):
        mean.append(train_x[:,i].mean())
        std.append(train_x[:,i].std())

    for i in range(train_x.shape[1]):
        for j in range(train_x.shape[0]):
            train_x[j,i] = (train_x[j,i]-mean[i])
            train_x[j,i] = (train_x[j,i]/std[i])

    for i in range(test_x.shape[1]):
        for j in range(test_x.shape[0]):
            test_x[j,i] = (test_x[j,i]-mean[i])
            test_x[j,i] = (test_x[j,i]/std[i])

    return (train_x,test_x)
```

##### Checking All Combinations of Features and trying to come up with best possible model.

```python
from itertools import combinations
from matplotlib import pyplot as mpl

def gradient_descent(X,y,meu,iter):
    no_of_rows = X.shape[0]
    no_of_columns = X.shape[1]
    W = np.ones((no_of_columns,)).reshape(-1,1) # making a column 2d vector from 1d vector
    y = y.reshape(-1,1) # making a column 2d vector from 1d vector

    temp_cost = (1/no_of_rows)*( (((X@W)-y).T) @ ((X@W)-y) )[0,0]
    old_cost = 9223372036854775807 # infinity
    cost = []
    cost.append(temp_cost)

    i=0
    while(i<=iter and old_cost-temp_cost >= 0.000001):
        i+=1
        
        W = W - meu*((1/no_of_rows)*((X.T) @ (X@W-y)))
        old_cost = temp_cost
        
        temp_cost = (1/no_of_rows)*( (((X@W)-y).T) @ ((X@W)-y) )[0,0]
        cost.append(temp_cost)
        
        if temp_cost > old_cost:
            return (W,-1)
    
    return (W,temp_cost)


def gradient_for_combination(normalised_train_x,y,comb,meu,iter):
    
    columns = list(comb)
        
    new_X_normalised = normalised_train_x[:,columns]

    ones_columns = np.ones((new_X_normalised.shape[0],)).reshape(-1,1)
    full_new_X = np.concatenate((ones_columns,new_X_normalised),axis=1)
    tup=gradient_descent(full_new_X,y,meu,iter)
    # print(tup)
    return tup

def best_combination(normalised_train_x,y):
    
    possibilities = []

    no_of_columns = normalised_train_x.shape[1]
    
    for i in range(no_of_columns):
        # they were taking a lot of time for the computation
        if(i==4 or i== 5 or i== 6):
            continue
        comb = combinations(range(no_of_columns), i+1)        
        for i in list(comb):
            if( 12 in i):
                if(13 not in i):
                    continue
            if( 13 in comb):
                if(12 not in comb):
                    continue
            possibilities.append(i)
    
    best_comb = possibilities[0]
    best_accuracy = gradient_for_combination(normalised_train_x,y,best_comb,0.001,6000)

    for i in possibilities:

        if(len(i) <= 2):
            temp = gradient_for_combination(normalised_train_x,y,i,0.001,6000)
        else:
            temp = gradient_for_combination(normalised_train_x,y,i,0.05,1000)

        if(temp[1] == -1):
            continue
        if temp[1]<best_accuracy[1]:
            best_accuracy = temp
            best_comb = i
    
    return (best_accuracy,best_comb)
    
```

#### Error Calculations

```python
# mean absolute relative error
mean_absolute_relative_error = lambda y, yhat, epsilon: (100/y.size)*np.sum((np.abs(y-yhat)/(np.abs(y) + epsilon)))
```

```python
def error_calc(train_x,test_x,train_y,test_y,columns):
    
    train_x_copy = np.ndarray.copy( train_x)
    test_x_copy = np.ndarray.copy( test_x)
    normalised_train_x,normalised_test_x = normalize(train_x_copy,test_x_copy)

    normalised_train_x = normalised_train_x[:,columns]
    normalised_test_x = normalised_test_x[:,columns]

    first_column = np.ones((normalised_train_x.shape[0],)).reshape(-1,1)
    normalised_train_x = np.concatenate((first_column,normalised_train_x),axis=1)
    
    first_column = np.ones((normalised_test_x.shape[0],)).reshape(-1,1)
    normalised_test_x = np.concatenate((first_column,normalised_test_x),axis=1)
    
    train_ypred = normalised_train_x @ W
    test_ypred = normalised_test_x @ W
    
    print("Training VS Testing\n")
    print("Mean Squared Error - " + str(mean_squared_error(train_y, train_ypred)) + " " + str(mean_squared_error(test_y, test_ypred)))
    print("Mean Absolute Error - " + str(mean_absolute_error(train_y, train_ypred)) + " " + str(mean_absolute_error(test_y, test_ypred)))
    train_ypred = np.reshape(train_ypred,-1)
    test_ypred = np.reshape(test_ypred,-1)
    print("Mean Absolute Relative Error - " + str(mean_absolute_relative_error(train_y, train_ypred,0.01)) + " " + str(mean_absolute_relative_error(test_y, test_ypred,0.01)))
```

##### Calculating W

```python
def closed_form(X,y):
    temp = X.T @ X
    W = (np.linalg.inv(temp) @ X.T) @ y
    return W
```

## Train Test Split

As the size of data is not very large we are using 60%(training) to 40%(testing) split as discussed in Tutorial session.

```python
# Separating the expected output from features

y = main_df["Class"].to_numpy()
main_df.drop(axis="columns", labels="Class", inplace=True)
X = main_df.to_numpy()

y_sec = main_df_copy["Class"].to_numpy()
main_df_copy.drop(axis="columns", labels="Class", inplace=True)
X_sec = main_df_copy.to_numpy()
```

```python
def train_test_split(X,y,training_size):

    np.random.seed(0)
    random_order = np.random.permutation(X.shape[0])

    # Below given is using random_order to randomly order the rows exactly in same order in both X and y
    X = X[random_order]
    y = y[random_order]

    division = int(training_size*X.shape[0])

    train_x = X[0:division,:]
    train_y = y[0:division]
    test_x = X[division:,:]
    test_y = y[division:]
    
    return (train_x,train_y,test_x,test_y)
```

```python
train_x,train_y,test_x,test_y = train_test_split(X,y,0.6) # class 2 or 4
train_x_sec,train_y_sec,test_x_sec,test_y_sec = train_test_split(X_sec,y_sec,0.6) # class 0 or 1
```

#### Normalisation of the data

```python
#function to normalise data before finding gradient
# not used -- was creating problems such as division by zero and here we can get accuracy without normalisation
def normalize(train_x,test_x):
    
    mean = []
    std = []

    for i in range(train_x.shape[1]):
        mean.append(train_x[:,i].mean())
        std.append(train_x[:,i].std())

    for i in range(train_x.shape[1]):
        for j in range(train_x.shape[0]):
            train_x[j,i] = (train_x[j,i]-mean[i])
            train_x[j,i] = (train_x[j,i]/std[i])

    for i in range(test_x.shape[1]):
        for j in range(test_x.shape[0]):
            test_x[j,i] = (test_x[j,i]-mean[i])
            test_x[j,i] = (test_x[j,i]/std[i])

    return (train_x,test_x)
```

```python
def normal(x,mean,sigma):
    return 1/(np.sqrt(2*np.pi)*sigma)*(np.exp(-1*0.5*((x-mean)/sigma)**2))
```

#### Naive Bayes Algo

```python
def naive_bayes(X_temp,mean_array_b_temp,std_array_b_temp,mean_array_m_temp,std_array_m_temp,column):
    
    X = X_temp[:,column]

    mean_array_b_temp = mean_array_b_temp.tolist()
    std_array_b_temp = std_array_b_temp.tolist()
    mean_array_m_temp = mean_array_m_temp.tolist()
    std_array_m_temp = std_array_m_temp.tolist()

    mean_array_b = []
    std_array_b = []
    mean_array_m = []
    std_array_m = []

    for i in column:
        mean_array_b.append(mean_array_b_temp[i])
        std_array_b.append(std_array_b_temp[i])
        mean_array_m.append(mean_array_m_temp[i])
        std_array_m.append(std_array_m_temp[i])

    y_pred = []

    q_b = 0
    q_m = 0

    for i in X:

        q_b = P_b
        k=0
        for j in i:
            q_b *= normal(j,mean_array_b[k],std_array_b[k])
            k+=1

        q_m = P_m
        k=0
        for j in i:
            q_m *= normal(j,mean_array_m[k],std_array_m[k])
            k+=1

        if(q_b > q_m):
            y_pred.append(2)
        else:
            y_pred.append(4)

    return y_pred

y_pred = naive_bayes(train_x,train_x_b_mean,train_x_b_std,train_x_m_mean,train_x_m_std,[0,1,2,3,4,5,6,7,8])
y_test_pred = naive_bayes(test_x,test_x_b_mean,test_x_b_std,test_x_m_mean,test_x_m_std,[0,1,2,3,4,5,6,7,8])

y_pred = np.array(y_pred)
y_test_pred = np.array(y_test_pred)

print(accuracy_score(train_y, y_pred))
print(accuracy_score(test_y, y_test_pred))
```

## Gradient Method

### Making a gradient vector

```python
def gradient_vector(X,W,y):
    column_of_sigmoid = np.exp((-1)*(X@W)) + 1
    column_of_sigmoid = 1/column_of_sigmoid
    temp = y-column_of_sigmoid
    result = X.T @ temp
    return result
```

### Finding Likelihood

```python
def likelihood_step(X,W,y):
    one = (-1)*(np.sum(np.log(np.exp(X@W) + 1)))
    two = (X@W).T @ y
    likelihood = one + two[0,0]
    return likelihood
```

### Gradient Algo

```python
def gradient_method(X,y_temp,meu):
    iter = 2000
    no_of_columns = X.shape[1]
    W = np.ones((no_of_columns,)).reshape(-1,1) # making a column 2d vector from 1d vector
    y = y_temp.reshape(-1,1) # making a column 2d vector from 1d vector

    temp_likelihood = likelihood_step(X,W,y)
    old_likelihood = -9223372036854775807 # infinity
    likelihoods = []
    likelihoods.append(temp_likelihood)

    i=0
    while(i<=iter and temp_likelihood-old_likelihood >= 0.000001):
        i+=1

        gradient_vector_column = gradient_vector(X,W,y)
        W = W + meu*gradient_vector_column
        
        old_likelihood = temp_likelihood
        temp_likelihood = likelihood_step(X,W,y)
        likelihoods.append(temp_likelihood)

        if temp_likelihood < old_likelihood:
            return (W,-1)
            
    return (W,temp_likelihood)
```

## Newtons Method

#### finding the inverse of Hessian Matrix

```python

def inv_H(X, W):
    
    temp = 1/(1+np.exp((-1)*(X@W)))
    y = np.reshape(temp,-1)
    
    S = np.diag(y*(1-y))
    
    H = X.T @ S @X
    return np.linalg.inv(H)
```

#### Newton's Method

```python
def newtons_method(X,y_temp,meu):
    iter = 2000
    no_of_columns = X.shape[1]
    W = np.ones((no_of_columns,)).reshape(-1,1) # making a column 2d vector from 1d vector
    y = y_temp.reshape(-1,1) # making a column 2d vector from 1d vector

    temp_likelihood = likelihood_step(X,W,y)
    old_likelihood = -9223372036854775807 # infinity
    likelihoods = []
    likelihoods.append(temp_likelihood)

    i=0
    while(i<=iter and temp_likelihood-old_likelihood >= 0.000001):
        i+=1

        gradient_vector_column = gradient_vector(X,W,y)
        W = W + inv_H(X,W)@gradient_vector_column
        
        old_likelihood = temp_likelihood
        temp_likelihood = likelihood_step(X,W,y)
        likelihoods.append(temp_likelihood)

        if temp_likelihood < old_likelihood:
            return (W,-1)
            
    return (W,temp_likelihood)
```
